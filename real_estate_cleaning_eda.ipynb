```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=df.drop_duplicates(
    subset=["area_type" , "location" , "Sahe"] )
```

```python
df
```

```python
sns.boxplot(df["price"])
```

```python
q1 = df["price"].quantile(0.25)
q3 = df["price"].quantile(0.75)
iqr = q3 - q1

low = q1 - 1.5 * iqr
high = q3 + 1.5 * iqr

outliers = df[(df["price"] < low) | (df["price"] > high)]
outliers
```

```python
df.drop(outliers.index, inplace=True)
```

```python
df
```

```python
sns.boxplot(df["price"])
```

```python
INR_TO_AZN = 0.0184689  

df["price_azn"] = (df["price"].astype(float) * 100000 * INR_TO_AZN).round(2)
```

```python
df
```

```python
df=df.drop(columns=["price"])
```

```python
df
```

```python
plt.figure(figsize=(8,4))
sns.histplot(df["price_azn"].dropna(), bins=60, kde=True)
plt.title("Price per m² (INR) - Distribution")
plt.show()
```

```python
top_locs = df["location"].value_counts().head(10).index
df_top = df[df["location"].isin(top_locs)].copy()

plt.figure(figsize=(12,5))
sns.boxplot(data=df_top, x="location", y="price_azn")
plt.xticks(rotation=90)
plt.title("Price per m² by Location (Top 10)")
plt.show()
```

```python
med = (df_top.groupby("location")["price_azn"]
       .median()
       .sort_values(ascending=False)
       .reset_index())

plt.figure(figsize=(10,6))
sns.barplot(data=med, x="price_azn", y="location")
plt.title("Median Price per m² by Location (Top 10)")
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.boxplot(data=df, x="area_type", y="price_azn")
plt.xticks(rotation=20)
plt.title("Price per m² by Area Type")
plt.show()
```

```python
num_cols = [ "price_azn", "Bedroom", "bath", "balcony"]
corr = df[num_cols].corr()

plt.figure(figsize=(8,6))
sns.heatmap(corr, annot=True, fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```
